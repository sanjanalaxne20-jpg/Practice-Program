```python
#find the lcm
num1=int(input("Enter first number:"))
num2=int(input("Enter Secound number:"))
greater=max(num1,num2)
while True:
    if greater % num1 == 0 and greater % num2==0:
        print("LCM=",greater)
        break
    greater +=1
```

```python
rows=5
num=1
for i in range(1,rows+1):
    for j in range(i):
        print(num,end=" ")
        num+=1
    print()
```
