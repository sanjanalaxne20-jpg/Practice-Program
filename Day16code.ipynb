```python
#Reverse Every Word
s = input("Enter a sentence: ")
words = s.split()
for word in words:
    print(word[::-1], end=" ")
```

```python
#Count Frequency of Each Character
s = input("Enter a string: ")
for ch in set(s):
    print(ch, "=", s.count(ch))
    
```

```python
#Remove Duplicate Characters
s = input("Enter a string: ")
result = ""
for ch in s:
    if ch not in result:
        result += ch
print("New String =", result)
```

```python
#Count Number of Words
s = input("Enter a sentence: ")
words = s.split()
print("Total Words =", len(words))
```

```python
#Find Largest Word
s = input("Enter a sentence: ")

words = s.split()

largest = words[0]

for word in words:
    if len(word) > len(largest):
        largest = word

print("Largest Word =", largest)
```

```python
#Find smallest Word
s = input("Enter a sentence: ")

words = s.split()

smallest = words[0]

for word in words:
    if len(word) < len(smallest):
        smallest = word

print("Smallest Word =", smallest)
```

```python
#Replace a Word
s = input("Enter a sentence: ")

old = input("Word to replace: ")
new = input("New word: ")

print(s.replace(old, new))
```

```python
#Count Uppercase and Lowercase Letters
s = input("Enter a string: ")
upper = lower = 0
for ch in s:
    if ch.isupper():
        upper += 1
    elif ch.islower():
        lower += 1
print("Uppercase =", upper)
print("Lowercase =", lower)
```

```python
#Remove All Spaces
s = input("Enter a string: ")
print(s.replace(" ", ""))
```
