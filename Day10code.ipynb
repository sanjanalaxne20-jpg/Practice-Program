```python
#create and print tuple
t=(1,2,3,4)
print(t)
```

```python
#find the length of tuple
t=(1,2,3,4,5)
print("Length:",len(t))
```

```python
#Access the tuple element
t=("Apple","banana","Mango")
print(t[0])
print(t[1])
```

```python
#Find the maximum and minimum value
t=(45,55,35,67,88)
print("Maximum:",max(t))
print("Minimum:",min(t))
```

```python
#sum of tuple element
t=(1,2,3,4,5)
print("Sum:",sum(t))
```

```python
#count an element
t=(1,2,3,4,5)
print(t.count(4))
```

```python
#find the index an element
t=(1,2,3,4,5,6)
print(t.index(5))
```

```python
#Check the whwther an element exist
t=(1,2,3,4,5,6,7,8,9)
if 7 in t:
    print("Element is Found  ")
else:
    print("Element is not Found ")
```

```python
#reverse tuple
t=(1,2,3,4,5,6,7,8,9)
print(t[::-1])
```

```python
#Convert tuple into list
t=(1,2,3,4,5,6,7,8,9)
l=list(t)
print(l)
```
