```python
#Find Length of List
def list_length(l1):
    return len(l1)

l1 = list(map(int, input("Enter numbers: ").split()))

print("Length =", list_length(l1))
```

```python
#Count an Element
def count_element(l1, x):
    return l1.count(x)

l1 = list(map(int, input("Enter numbers: ").split()))
x = int(input("Enter element to count: "))

print("Count =", count_element(l1, x))
```

```python
#Find Index of an Element
def find_index(l1, x):
    return l1.index(x)

l1 = list(map(int, input("Enter numbers: ").split()))
x = int(input("Enter element: "))

print("Index =", find_index(l1, x))
```

```python
#Append an Element
def add_element(l1, x):
    l1.append(x)
    return l1

l1 = list(map(int, input("Enter numbers: ").split()))
x = int(input("Enter element to append: "))

print(add_element(l1, x))
```

```python
#Insert an Element
def insert_element(l1, pos, value):
    l1.insert(pos, value)
    return l1

l1 = list(map(int, input("Enter numbers: ").split()))
pos = int(input("Enter position: "))
value = int(input("Enter value: "))

print(insert_element(l1, pos, value))
```

```python
#Remove an Element
def remove_element(l1, x):
    l1.remove(x)
    return l1

l1 = list(map(int, input("Enter numbers: ").split()))
x = int(input("Enter element to remove: "))

print(remove_element(l1, x))
```

```python
#Pop Last Element
def pop_element(l1):
    l1.pop()
    return l1

l1 = list(map(int, input("Enter numbers: ").split()))

print(pop_element(l1))
```

```python
#Extend Two Lists
def extend_list(l1, l2):
    l1.extend(l2)
    return l1

l1 = list(map(int, input("Enter first list: ").split()))
l2 = list(map(int, input("Enter second list: ").split()))

print(extend_list(l1, l2))
```

```python
#Copy a List
def copy_list(l1):
    return l1.copy()

l1 = list(map(int, input("Enter numbers: ").split()))

print(copy_list(l1))
```

```python
#Clear a List
def clear_list(l1):
    l1.clear()
    return l1

l1 = list(map(int, input("Enter numbers: ").split()))

print(clear_list(l1))
```
