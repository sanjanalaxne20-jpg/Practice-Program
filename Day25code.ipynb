```python
#Create a function to calculate electricity bill.
def electricity_bill(units):
    if units <= 100:
        bill = units * 5
    elif units <= 200:
        bill = units * 7
    else:
        bill = units * 10       
    return bill
print("Bill =", electricity_bill(250))
```

```python
#Salary Calculator
#Problem:
def salary(basic):
    hra = basic * 0.20
    da = basic * 0.10
    
    total = basic + hra + da
    
    return total

print("Salary =", salary(30000))
```

```python
#Shopping Discount
#Problem:
def discount(amount):
    if amount >= 5000:
        return amount * 0.20
    elif amount >= 2000:
        return amount * 0.10
    else:
        return 0
print("Discount =", discount(6000))
```

```python
#Bank Interest Calculator
#Problem:
def interest(p, r, t):
    return (p*r*t)/100

print(interest(10000,5,2))
```

```python
#Age Calculator
#Problem:
def age(birth_year):
    current_year = 2026
    return current_year - birth_year

print("Age =", age(2004))
```

```python
#Taxi Fare Calculator
#Problem:
def taxi_fare(distance):
    fare = distance * 15
    return fare

print("Fare =", taxi_fare(10))
```

```python
#Employee Bonus Calculator
#Problem:
def bonus(salary):
    if salary > 50000:
        return salary*0.10
    else:
        return salary*0.05

print(bonus(60000))
```

```python
#BMI Checker
#Problem:
def bmi(weight, height):
    result = weight/(height**2)
    
    return result

print(bmi(70,1.75))
```

```python
#ATM Withdrawal
#Problem:
def withdraw(balance, amount):
    if amount <= balance:
        return balance - amount
    else:
        return "Insufficient Balance"

print(withdraw(10000,3000))
```

```python
#Student Percentage Calculator
#Problem:
def percentage(m1, m2, m3, m4, m5):
    total = m1+m2+m3+m4+m5
    per = (total/500)*100
    return per

print("Percentage =", percentage(80,75,90,85,70))
```
