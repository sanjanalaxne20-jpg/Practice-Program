```python
#Hotel Bill
def hotel_bill(days, rent):
    return days * rent

print(hotel_bill(5, 1500))
```

```python
#Courier Charge
def courier(weight):
    return weight * 50

print(courier(4))
```

```python
#Water Bill
def water_bill(units):
    return units * 12

print(water_bill(35))
```

```python
#Internet Bill
def internet_bill(months):
    return months * 799

print(internet_bill(3))
```

```python
#Loan EMI (Simple)
def emi(amount, months):
    return amount / months

print(emi(120000, 12))
```

```python
#Shopping GST
def gst(amount):
    return amount + (amount * 18 / 100)

print(gst(2500))
```

```python
#Employee Overtime Pay
def overtime(hours, rate):
    return hours * rate

print(overtime(12, 250))
```

```python
#Calculate Commission
def commission(sales):
    return sales * 0.05

print(commission(50000))
```

```python
#Library Fine
def library_fine(days):
    return days * 2

print(library_fine(8))
```

```python
#Train Ticket Fare
def train_fare(passengers, fare):
    return passengers * fare

print(train_fare(3, 450))
```
