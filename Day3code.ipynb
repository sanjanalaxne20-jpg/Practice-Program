```python
#write a program factorial of user given number
num=int(input("enter the number"))
res=1
i=1
while(i<=num):
    res=res*i
    i=i+1
print(res)
```

```python
#REVERSE FACTORIAL
num=int(input("enter the number"))
res1=1
while(num>0):
    res1=res1*num
    num=num-1
print(res1)
```

```python
#write a program to print vowels in words
word=str(input("Enter the words"))
for vowel in word:
    if vowel=='a'or vowel=='e'or vowel=='i'or vowel=='o'or vowel=='u':
        print(vowel)
```
