```python
# Creating a dictionary
student = {"Name": "Zzz", "Age": 20, "Course": "B.tech"}
print(student)
```

```python
#Access a Value
student = {"Name": "Zzz", "Age": 20}
print(student["Name"])
```

```python
#Add a New Key-Value Pair
student = {"Name": "Zeba"}
student["Age"] = 20
print(student)
```

```python
#update a value
student = {"Name": "Zeba", "Age": 20}
student["Age"] = 21
print(student)
```

```python
#Delete key
student = {"Name": "Zeba", "Age": 20}
del student["Age"]
print(student)
```

```python
#Find the Length of a Dictionary
student = {"A": 1, "B": 2, "C": 3}
print(len(student))
```

```python
#Print All Keys
student = {"Name": "Zeba", "Age": 20}
print(student.keys())
```

```python
#Print all value
student = {"Name": "Zeba", "Age": 20}
print(student.values())
```

```python
#Print all items
student = {"Name": "Zeba", "Age": 20}
print(student.items())
```

```python
#Check if a Key Exists
student = {"Name": "Zeba", "Age": 20}
if "Age" in student:
    print("Key Found")
else:
    print("Key Not Found")
```

```python
#clear dictionary
student = {"Name": "Zeba", "Age": 20}
student.clear()
print(student)
```
