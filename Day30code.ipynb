```python
#Increase Every Element by 10
l1 = list(map(int, input("Enter numbers: ").split()))
result = list(map(lambda x: x + 10, l1))
print(result)
```

```python
#Double Every Element
l1 = list(map(int, input("Enter numbers: ").split()))
result = list(map(lambda x: x * 2, l1))
print(result)
```

```python
#Square Every Element
l1 = list(map(int, input("Enter numbers: ").split()))
result = list(map(lambda x: x ** 2, l1))
print(result)
```

```python
#Convert String List to Uppercase
l1 = input("Enter names: ").split()
result = list(map(lambda x: x.upper(), l1))
print(result)
```

```python
#Add 5 to Every Number
l1 = list(map(int, input("Enter numbers: ").split()))
result = list(map(lambda x: x + 5, l1))
print(result)
```

```python
#Convert to Lowercase
l1 = input("Enter words: ").split()
result = list(map(lambda x: x.lower(), l1))
print(result)
```

```python
#Remove Spaces
text = input("Enter text: ")
result = (lambda x: x.replace(" ", ""))(text)
print(result)
```

```python
#Multiply by 3
l1 = list(map(int, input("Enter numbers: ").split()))
result = list(map(lambda x: x * 3, l1))
print(result)
```

```python
#Round Float Numbers
l1 = list(map(float, input("Enter numbers: ").split()))
result = list(map(lambda x: round(x), l1))
print(result)
```

```python
#Convert Celsius to Fahrenheit
l1 = list(map(int, input("Enter temperatures: ").split()))
result = list(map(lambda c: (c * 9/5) + 32, l1))
print(result)
```
