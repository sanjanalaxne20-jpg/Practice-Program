```python
#Check Perfect Number
num = int(input("Enter a number: "))

total = 0

for i in range(1, num):
    if num % i == 0:
        total += i

if total == num:
    print("Perfect Number")
else:
    print("Not Perfect Number")
```

```python
#Strong Number
import math

num = int(input("Enter a number: "))

temp = num
total = 0

while temp > 0:
    digit = temp % 10
    total += math.factorial(digit)
    temp //= 10

if total == num:
    print("Strong Number")
else:
    print("Not Strong Number")
```

```python
#Neon Number
num = int(input("Enter a number: "))

square = num * num
total = 0

while square > 0:
    total += square % 10
    square //= 10

if total == num:
    print("Neon Number")
else:
    print("Not Neon Number")
```

```python
#Spy Number
num = int(input("Enter a number: "))

temp = num
s = 0
p = 1

while temp > 0:
    digit = temp % 10
    s += digit
    p *= digit
    temp //= 10

if s == p:
    print("Spy Number")
else:
    print("Not Spy Number")

```

```python
#HCF of Two Numbers
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

small = min(a, b)

for i in range(small, 0, -1):
    if a % i == 0 and b % i == 0:
        print("HCF =", i)
        break
```

```python
#LCM of Two Numbers
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

large = max(a, b)

while True:
    if large % a == 0 and large % b == 0:
        print("LCM =", large)
        break
    large += 1
```

```python
#Count Even and Odd Digits
num = input("Enter a number: ")

even = odd = 0

for i in num:
    if int(i) % 2 == 0:
        even += 1
    else:
        odd += 1

print("Even Digits =", even)
print("Odd Digits =", odd)
```

```python
#Print Factors
num = int(input("Enter a number: "))

for i in range(1, num + 1):
    if num % i == 0:
        print(i)
```

```python
#Remove All Vowels
text = input("Enter text: ")

result = ""

for i in text:
    if i.lower() not in "aeiou":
        result += i

print(result)
```

```python
#Check Binary Number
num = input("Enter a binary number: ")

if set(num) <= {"0", "1"}:
    print("Valid Binary")
else:
    print("Invalid Binary")
```
