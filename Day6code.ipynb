```python
#Palidrome Number
num=int(input("Enter a number:"))
temp=num
rev=0
while temp>0:
    digit=temp%10
    rev=rev*10+digit
    temp=temp//10
if num == rev:
    print("palidrome number")
else:
    print("Not a Palidrone Number")
```

```python
#Armstrong Number
num=153
temp=num
sum=0
while temp>0:
    digit=temp%10
    sum+=digit**3
    temp//=10
if sum == num:
    print("Armstrong")
else:
    print("Not Armstrong")
```

```python
#factorial
num=5
fact=1
for i in range(1,num+1):
    fact*=i
print(fact)
```

```python
#Fibonacci Series
a=0
b=1
for i in range(10):
    print(a, end=" ")
    a,b=b,a+b
```
