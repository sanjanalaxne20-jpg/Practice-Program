```python
#wap to add number using def function
def add(n1,n2):
    res=n1+n2
    return res
total = add(10,20)
print(total)
```

```python
#wap to substract number using def function
def sub(n1,n2):
    res=n1-n2
    return res
total = sub(20,10)
print(total)
```

```python
#wap to Multiplication number using def function
def mul(n1,n2):
    res=n1*n2
    return res
total = mul(20,10)
print(total)
```

```python
#wap to division number using def function
def div(n1,n2):
    res=n1/n2
    return res
total = div(20,10)
print(total)
```

```python
#wap for floor division number using def function
def mod(n1,n2):
    res=n1//n2
    return res
total = mod(20,10)
print(total)
```

```python
#wap to remainder number using def function
def rem(n1,n2):
    res=n1%n2
    return res
total = rem(20,10)
print(total)
```

```python
#wap to power using def function
def power(n1,n2):
    res=n1**n2
    return res
total = power(20,2)
print(total)
```

```python
#wap to cube root using def function
def cube_root(n1):
    res=n1*1//3
    return res
total = cube_root(8)
print(total)
```

```python
#wap to cube root using def function
def square_root(n1):
    res=n1*1//2
    return res
total = square_root(8)
print(total)
```
