```python
for i in range(1,5):
    if i%2==0:
        print('#'*i)
    else:
        print('&'*i)
```

```python
for i in range(1,11):
    if i==5:
        break
    print(i)
else:
    print("Hello")
```

```python
for i in range(1,11):
    if i==5 or i==3 or i==7:
        continue
    print(i)
else:
    print("Hello")
```

```python
#wap to check user given number is prime or not
num=int(input("Enter the number"))
for i in range(2,num):
    if num%i==0:
        print("Not prime")
        break
else:
    print("prime")
```

```python
#wap on pass keyword
for i in range(1,11):
    if i==5:
        pass
    print(i)
else:
    print("Hello")
```

```python
num=int(input("Enter the number"))
for i in range(1,num):
    for j in range(1,i+1):
        print(j, end=" ")
    print()
```

```python
num=int(input("Enter the number"))
for i in range(num,0,-1):
    for j in range(1,i+1):
        print(j, end=" ")
    print()
```

```python
num=int(input("Enter the number"))
for i in range(1,num+1):
    print(" "*(num-i)+"*"*(2*i-1))
for i in range(num-1,0,-1):
    print(" "*(num-i)+"*"*(2*i-1))
```
