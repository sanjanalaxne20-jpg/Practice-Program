```python
#Removes and returns a random element.(pop)
s = {10, 20, 30}
print(s.pop())
print(s)
```

```python
#Removes all elements.clear()
s = {10, 20, 30}
s.clear()
print(s)
```

```python
#Creates a copy of the set.copy()
s = {10, 20, 30}
s.copy()
print(s)
```

```python
#Combines two sets.union()
A = {1, 2, 3}
B = {3, 4, 5}
print(A.union(B))
```

```python
#Returns common elements.intersection()
A = {1, 2, 3}
B = {2, 3, 4}
print(A.intersection(B))
```

```python
#Returns elements in the first set but not in the second.difference()
A = {1, 2, 3}
B = {2, 3, 4}
print(A.difference(B))
```

```python
#symmetric_difference()
A = {1, 2, 3}
B = {2, 3, 4}
print(A.symmetric_difference(B))
```

```python
#issubset()Checks whether one set is a superset of another.
A = {1, 2, 3}
B = {1, 2}
print(A.issuperset(B))
```

```python
#isdisjoint()Checks whether two sets have no common elements.
A = {1, 2}
B = {3, 4}
print(A.isdisjoint(B))
```

```python
s = {10, 20, 30, 40}
print(len(s))      # Number of elements
print(max(s))      # Largest element
print(min(s))      # Smallest element
print(sum(s))      # Sum of elements
print(sorted(s))   # Sorted list
print(type(s))     # Data type
```
