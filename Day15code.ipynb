```python
#Check if a Key Exists
student = {"Name": "Zeba", "Age": 20}
if "Age" in student:
    print("Key Found")
else:
    print("Key Not Found")
```

```python
#Copy a Dictionary
d1 = {"A": 10, "B": 20}
d2 = d1.copy()
print(d2)
```

```python
#Clear the Dictionary
student = {"Name": "Zeba", "Age": 20}
student.clear()
print(student)
```

```python
#Remove a Key Using pop()
student = {"Name": "Zeba", "Age": 20}
student.pop("Age")
print(student)
```

```python
#Merge Two Dictionaries
d1 = {"A": 1}
d2 = {"B": 2}
d1.update(d2)
print(d1)
```

```python
#Print All Keys Using a Loop
student = {"Name": "Zeba", "Age": 20}
for key in student:
    print(key)
```

```python
#Print All Values Using a Loop
student = {"Name": "Zeba", "Age": 20}
for value in student.values():
    print(value)
```

```python
#Print Key-Value Pairs
student = {"Name": "Zeba", "Age": 20}
for key, value in student.items():
    print(key, ":", value)
```

```python
#Create a Dictionary from Two Lists
keys = ["Name", "Age", "City"]
values = ["Zeba", 20, "Pune"]
d = dict(zip(keys, values))
print(d)
```

```python
#Find the Maximum Value
marks = {"Math": 85, "Science": 90, "English": 95}
print(max(marks.values()))
```

```python
#Find the Minimum Value
marks = {"Math": 85, "Science": 90, "English": 95}
print(min(marks.values()))
```
