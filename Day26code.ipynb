```python
#Movie Ticket Price
def ticket_price(persons):
    return persons * 250

print(ticket_price(4))
```

```python
#Pizza Bill
def pizza_bill(price, quantity):
    return price * quantity

print(pizza_bill(300, 3))
```

```python
#Mobile Recharge
def recharge(balance, recharge_amount):
    return balance + recharge_amount

print(recharge(100, 299))
```

```python
#Fuel Cost
def fuel_cost(liters, price):
    return liters * price

print(fuel_cost(20, 105))
```

```python
#Exam Result
def result(marks):
    if marks >= 35:
        return "Pass"
    else:
        return "Fail"

print(result(45))
```

```python
#Grade Calculator
def grade(marks):
    if marks >= 90:
        return "A"
    elif marks >= 75:
        return "B"
    elif marks >= 60:
        return "C"
    else:
        return "D"

print(grade(82))
```

```python
#Voting Eligibility
def voting(age):
    if age >= 18:
        return "Eligible"
    else:
        return "Not Eligible"

print(voting(20))
```

```python
#Driving License Eligibility
def license(age):
    if age >= 18:
        return "Can Apply"
    else:
        return "Cannot Apply"

print(license(17))
```

```python
#Cricket Strike Rate
def strike_rate(runs, balls):
    return (runs / balls) * 100

print(strike_rate(75, 50))
```

```python
#Batting Average
def batting_average(runs, innings):
    return runs / innings

print(batting_average(850, 20))
```
