```python
#Function with Default Argument
def greet(name="Student"):
    print("Hello", name)

greet()
greet("Sanjana")
```

```python
#Function with Keyword Arguments
def student(name, age):
    print("Name:", name)
    print("Age:", age)

student(age=21, name="Rahul")
```

```python
#Function with Variable-Length Arguments (*args)
def total(*numbers):
    print("Sum =", sum(numbers))

total(10, 20)
total(5, 10, 15, 20)
```

```python
#Function with Keyword Variable Arguments (**kwargs)
def details(**data):
    for key, value in data.items():
        print(key, ":", value)

details(Name="Riya", Age=20, City="Nagpur")
```

```python
#Lambda Function
square = lambda x: x * x

num = int(input("Enter number: "))
print("Square =", square(num))
```

```python
#Recursive Function (Factorial)
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)

num = int(input("Enter number: "))
print(factorial(num))
```

```python
def outer():
    print("Outer Function")

    def inner():
        print("Inner Function")

    inner()

outer()
```

```python
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)

n = int(input("Enter term: "))
print(fibonacci(n))
```

```python
#Function Returning Multiple Values
def calculation(a, b):
    return a + b, a - b, a * b

x = int(input("Enter first number: "))
y = int(input("Enter second number: "))

s, d, m = calculation(x, y)

print("Sum =", s)
print("Difference =", d)
print("Product =", m)
```

```python
#Function Using map()
def square(x):
    return x * x

numbers = [1, 2, 3, 4, 5]
result = list(map(square, numbers))

print(result)
```
