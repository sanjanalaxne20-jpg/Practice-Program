```python
#area of circle
num1=int(input("Enter the radius"))
area_of_circle=3.14*num1*num1
print(f"Area of Circle is {area_of_circle}")
```

```python
#Area of rectangle
num1=int(input("Enter the Length"))
num2=int(input("Enter the Breadth"))
num3=int(input("Enter the Height"))
area_of_rectangle=num1*num2*num3
print(f"Area of rectangle is {area_of_rectangle}")
```

```python
#Area of triangle
num1=int(input("Enter the Length"))
num2=int(input("Enter the Breadth"))
area_of_triangle=1/2*num1*num2
print(f"Area of triangle is {area_of_triangle}")
```
