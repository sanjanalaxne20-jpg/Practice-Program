```python
#addition of two number using lambda function
res=lambda a,b:a+b
res(3,5)
```

```python
#subtraction of two number using lambda function
res=lambda a,b:a-b
res(5,3)
```

```python
#multiplication of two number using lambda function
res=lambda a,b:a*b
res(5,3)
```

```python
#division of two number using lambda function
res=lambda a,b:a/b
res(2,1)
```

```python
#Remainder of two number using lambda function
res=lambda a,b:a%b
res(5,3)
```

```python
#finding square root using lambda function
sqrt=lambda x:x**(1/2)
sqrt(2)
```

```python
#fimdimg the cube root using lambda function
cbrt=lambda y:y**(1/3)
cbrt(81)
```

```python
#Area of side square using lambda function
SA=lambda s:s**2
SA(4)
```

```python
#find area of rectangle using lambda function
ra=lambda l,b,h:l*b*h
ra(2,3,4)
```

```python
#find the greater number using lambda function
gn=lambda a,b: a if a>b else b
gn(23,4)
```
