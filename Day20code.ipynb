```python
l1=[2,5,7,8,9,3,4,55,6,7]
def EV(x):
    if x%2==0:
        return True
    else:
        return False
```

```python
filter(EV,l1)
```

```python
list(filter(EV,l1))
```

```python
list(filter(lambda x: x%2==0,l1))
```

```python
#add two numbers
add = lambda a, b: a + b
print(add(10, 20))
```

```python
#check even or odd functions
even = lambda x: "Even" if x % 2 == 0 else "Odd"
print(even(17))
```

```python
#find the minimum using lambda function
minimum = lambda a, b: a if a < b else b
print(minimum(25, 18))
```

```python
#find the maximum using lambda function
maximum = lambda a, b: a if a>b else b
print(maximum(25,18))
```

```python
#find the largest of three numbers
largest = lambda a, b, c: max(a, b, c)
print("The largest number is :",largest(10, 30, 20))
```

```python
#find the smallest of three numbers
smallest=lambda a,b,c:min(a,b,c)
print("The smallest number is:",smallest(33.78,5,666.6))
```

```python
#convert the uppercase using lambda functions
upper = lambda s: s.upper()
print(upper("python"))
```

```python
#convert the lowercase using the lambda function
lower=lambda a:a.lower()
print(lower("FINISHED"))
```
