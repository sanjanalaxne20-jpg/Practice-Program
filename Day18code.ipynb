```python
#wap for area of rectangle using def function
def area_of_rectangle(n1,n2):
    res=n1*n2
    return res
total = area_of_rectangle(2,3)
print(total)
```

```python
#wap for area of sqaure using def function(n1=side)
def area_of_square(n1):
    res=n1**2
    return res
total = area_of_square(4)
print(total)
```

```python
#wap to area  of triangle using def function
def area_of_triangle(b,h):
    res=1/2*b*h
    return res
total = area_of_triangle(3,5)
print(total)
```

```python
#wap area of circle using def function
def area_of_circle(radius):
    res=3.14*radius*2
    return res
total = area_of_circle(3)
print(total)
```

```python
#find the maximum number using def function
def maximum(a,b):
    if a>b:
        return a
    else:
        return b
print(maximum(22,41))
```

```python
#find the minumun number using def function
s1=int(input("Enter the maximum number : "))
s2=int(input("Enter the maximum number : "))
def minimum(a,b):
    if a<b:
        print(f"{s1} is the minimum")
        return a
    else:
        print(f"{s2} is the minimum")
        return b
print(minimum(s1,s2))
```

```python
#find the even odd using def functions
def even_odd(num):
    if num%2==0:
        print("even")
    else:
        print("odd")
print(even_odd(3))
```

```python
#factorial find
def factorial(n):
    fact=1
    for i in range(1,n+1):
        fact*=i
    return fact
print(factorial(8))
        
```
