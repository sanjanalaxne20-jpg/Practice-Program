```python
#Append element
numbers=[10,20,30]
numbers.append(40)
print(numbers)
```

```python
#remove an element
numbers=[10,20,30,40]
numbers.remove(20)
print(numbers)
```

```python
#pop the last element
numbers=[10,20,30,40]
numbers.pop()
print(numbers)
```

```python
#count occurances of an element
numbers=[10,20,10,30,10]
print(numbers.count(10))
```

```python
#check if an element exists
numbers=[10,20,30,40]
if 20 in numbers:
    print("element found")
else:
    print("element not found")
```

```python
#Square of each elemnt
numbers=[1,2,3,4,5]
for i in numbers:
    print(i*i)
```

```python
#Print only odd numbers in list
numbers=[1,2,3,4,5,6]
for i in numbers:
    if i%2!=0:
        print(i)
```

```python
#find the frequency of each element
numbers=[1,2,2,3,3,3]
for i in set(numbers):
    print(i,":",numbers.count(i))
```

```python
#plus pattern
n=5
for i in range(n):
    for j in range(n):
        if i==n//2 or j==n//2:
            print("+",end=" ")
        else:
            print(" ",end=" ")
    print()
```

```python
#Butterfly pattern
n=5
for i in range(1,n+1):
    print("*"*i + " "*(2*(n-i)) + "*"*i)
for i in range(n,0,-1):
    print("*"*i + " "*(2*(n-i)) + "*"*i)
        
```
