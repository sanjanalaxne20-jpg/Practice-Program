```python
#wap right aligned inverted trianglr(*program)
num=int(input("Enter the number"))
for i in range(num,0,-1):
    print(" "*(num-i)+"*"*i)
```

```python
#numbers square pattern
num=int(input("Enter the number"))
for i in range(num):
    for j in range(1,num+1):
        print(j, end=" ")
    print()
```

```python
#Hollow square pattern
num=int(input("Enter the number"))
for i in range(num):
    for j in range(num):
        if i==0 or i==num-1 or j==0 or j==num-1:
            print(i,end=" ")
        else:
            print(" ", end=" ")
    print()
```

```python
#X pattern
num=int(input("Enter the number"))
for i in range(num):
    for j in range(num):
        if i==j or i+j==num-1 :
            print("*",end=" ")
        else:
            print(" ", end=" ")
    print()
```
